# Skew e Curtose
- Skew √© uma medida da simetria em uma distribui√ß√£o. Um conjunto de dados sim√©trico ter√° skew igual a 0. Assim, uma distribui√ß√£o normal ter√° skew de 0. 
- A medida curtose=0 indica mesmo achatamento da distribui√ß√£o normal. 
- A medida curtose>3 indica uma distribui√ß√£o mais concentrada que a normal. Com calda mais que a normal.

```python
print("Skew: %f" % data['SalePrice'].skew())
print("Kurtose: %f" % data['SalePrice'].kurt())
```

# Estudo de algumas vari√°veis que podem influenciar no pre√ßo da casa

Nosso objetivo aqui √© analisar a correla√ß√£o entre cada uma das vari√°veis empiricamente selecionadas, e a vari√°vel SalePrice. Dessa forma, foram escolhidas:

- OverallQual: atributo categ√≥rico que indica a qualidade do material e do acabamento
- GrLivArea: √°rea de estar da casa
- YearBuilt: data de constru√ß√£o da casa
- OverallCond: classifica√ß√£o da condi√ß√£o geral da casa
- TotalBsmtSF: √°rea do por√£o da casa
- LotFrontage: n√∫mero de p√©s (metros) da rua √† propriedade 
- EnclosedPorch: √Årea de varanda fechada em p√©s quadrados
- MoSold: m√™s de venda
- SalePrice: valor da casa


```python
#correlation matrix
new = data[['OverallQual', 'GrLivArea', 'YearBuilt', 
            'OverallCond', 'TotalBsmtSF', 'LotFrontage', 'EnclosedPorch', 'MoSold', 'SalePrice']].copy()

corrmat = new.corr()
f, ax = plt.subplots(figsize=(15, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

- Pela matrix, OverallCond, 'EnclosedPorch' e 'MoSold' n√£o tem correla√ß√£o com SalePrice
- J√° OverallQual, GrLivArea, TotalBsmSF aparentemente tem alta associa√ß√£o. 

# Coeficiente de correla√ß√£o
- O coeficiente de correla√ß√£o mostra qu√£o bem um modelo linear descreve a rela√ß√£o entre duas vari√°veis.
- Quando r=0, indica que as vari√°veis n√£o s√£o correlacionadas

```python
print("Coeficiente de correla√ß√£o r entre SalePrice e OverallCond: %f" % data.SalePrice.corr(data.OverallCond))
print("Coeficiente de correla√ß√£o r entre SalePrice e EnclosedPorch: %f" % data.SalePrice.corr(data.EnclosedPorch))
print("Coeficiente de correla√ß√£o r entre SalePrice e MoSold: %f" % data.SalePrice.corr(data.MoSold))
```

```python
plt.scatter(data.OverallCond, data.SalePrice)
```

```python
plt.scatter(data.EnclosedPorch, data.SalePrice)
```

```python
plt.scatter(data.MoSold, data.SalePrice)
```

- Quando r=1, indica que um modelo linear pode descrever a rela√ß√£o entre as vari√°veis


```python
print("Coeficiente de correla√ß√£o r entre SalePrice e GrLivArea: %f" % data.SalePrice.corr(data.GrLivArea))
print("Coeficiente de correla√ß√£o r entre SalePrice e TotalBsmtSF: %f" % data.SalePrice.corr(data.TotalBsmtSF))
print("Coeficiente de correla√ß√£o r entre SalePrice e OverallQual: %f" % data.SalePrice.corr(data.OverallQual))
print("Coeficiente de correla√ß√£o r entre SalePrice e YearBuilt: %f" % data.SalePrice.corr(data.YearBuilt))
```
