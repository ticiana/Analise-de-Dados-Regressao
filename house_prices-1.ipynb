# Skew e Curtose
- Skew é uma medida da simetria em uma distribuição. Um conjunto de dados simétrico terá skew igual a 0. Assim, uma distribuição normal terá skew de 0. 
- A medida curtose=0 indica mesmo achatamento da distribuição normal. 
- A medida curtose>3 indica uma distribuição mais concentrada que a normal. Com calda mais que a normal.

```python
print("Skew: %f" % data['SalePrice'].skew())
print("Kurtose: %f" % data['SalePrice'].kurt())
```

# Estudo de algumas variáveis que podem influenciar no preço da casa

Nosso objetivo aqui é analisar a correlação entre cada uma das variáveis empiricamente selecionadas, e a variável SalePrice. Dessa forma, foram escolhidas:

- OverallQual: atributo categórico que indica a qualidade do material e do acabamento
- GrLivArea: área de estar da casa
- YearBuilt: data de construção da casa
- OverallCond: classificação da condição geral da casa
- TotalBsmtSF: área do porão da casa
- LotFrontage: número de pés (metros) da rua à propriedade 
- EnclosedPorch: Área de varanda fechada em pés quadrados
- MoSold: mês de venda
- SalePrice: valor da casa


```python
#correlation matrix
new = data[['OverallQual', 'GrLivArea', 'YearBuilt', 
            'OverallCond', 'TotalBsmtSF', 'LotFrontage', 'EnclosedPorch', 'MoSold', 'SalePrice']].copy()

corrmat = new.corr()
f, ax = plt.subplots(figsize=(15, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

- Pela matrix, OverallCond, 'EnclosedPorch' e 'MoSold' não tem correlação com SalePrice
- Já OverallQual, GrLivArea, TotalBsmSF aparentemente tem alta associação. 

# Coeficiente de correlação
- O coeficiente de correlação mostra quão bem um modelo linear descreve a relação entre duas variáveis.
- Quando r=0, indica que as variáveis não são correlacionadas

```python
print("Coeficiente de correlação r entre SalePrice e OverallCond: %f" % data.SalePrice.corr(data.OverallCond))
print("Coeficiente de correlação r entre SalePrice e EnclosedPorch: %f" % data.SalePrice.corr(data.EnclosedPorch))
print("Coeficiente de correlação r entre SalePrice e MoSold: %f" % data.SalePrice.corr(data.MoSold))
```

```python
plt.scatter(data.OverallCond, data.SalePrice)
```

```python
plt.scatter(data.EnclosedPorch, data.SalePrice)
```

```python
plt.scatter(data.MoSold, data.SalePrice)
```

- Quando r=1, indica que um modelo linear pode descrever a relação entre as variáveis


```python
print("Coeficiente de correlação r entre SalePrice e GrLivArea: %f" % data.SalePrice.corr(data.GrLivArea))
print("Coeficiente de correlação r entre SalePrice e TotalBsmtSF: %f" % data.SalePrice.corr(data.TotalBsmtSF))
print("Coeficiente de correlação r entre SalePrice e OverallQual: %f" % data.SalePrice.corr(data.OverallQual))
print("Coeficiente de correlação r entre SalePrice e YearBuilt: %f" % data.SalePrice.corr(data.YearBuilt))
```

- YearBuilt aparentemente é também linearmente associado a SalePrice, porém não tão bem como os outros 
atributos analisados.
